### Data Preprocessing

```python
print("Supplies Subgroup: ",data1['Supplies Subgroup'].unique() )
print("Region : ",data1['Region'].unique())
print("Route To Market : ",data1['Route To Market'].unique())
print("Opportunity Result : ",data1['Opportunity Result'].unique())
print("Competitor Type : ",data1['Competitor Type'].unique())
print("'Supplies Group : ",data1['Supplies Group'].unique())

#Before leveling the data checking for the unique string values within each column of the dataframe
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data1['Supplies Subgroup'] = le.fit_transform(data1['Supplies Subgroup'])
data1['Region'] = le.fit_transform(data1['Region'])
data1['Route To Market'] = le.fit_transform(data1['Route To Market'])
data1['Opportunity Result'] = le.fit_transform(data1['Opportunity Result'])
data1['Competitor Type'] = le.fit_transform(data1['Competitor Type'])
data1['Supplies Group'] = le.fit_transform(data1['Supplies Group'])
data1.head()

#Converting String based data into numeric value using LabelEncoder() method from sklearn preprocessing module
```

### Training Set and Test Set

```python
cols = [col for col in data1.columns if col not in ['Opportunity Number','Opportunity Result']]
#Dropped 'Opportunity Number','Opportunity Result' columns from the DataFrame 
data2 = data1[cols] #Create new DataFrame after dropping 'Opportunity Number','Opportunity Result' and this is the Feature
target = data1['Opportunity Result'] #Set Opportunity Result as target variables and created new DataFrame target
data2.head(5)
```

```python
#Divide target and feature data frames into training set and testing sets with 70% training data and 30% testing data
from sklearn.model_selection import train_test_split
data2_train, data2_test, target_train, target_test = train_test_split(data2,target, test_size = 0.30, random_state = 10)
```

### Building Prediction Model
#### Naive Bayes Algorithm

```python
# Using Naive Bayes Algorithm
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gnb = GaussianNB() #Create an object of type GaussianNB
pred = gnb.fit(data2_train,target_train).predict(data2_test) # Train the algorithm on training data and predict using the test data
print("Naive-Bayes accuracy:",accuracy_score(target_test,pred,normalize = True))
```

### Linear SVC (Linear Support Vector Classification)

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
svc_model = LinearSVC(random_state = 0)
pred = svc_model.fit(data2_train,target_train).predict(data2_test)
print("LinearSVC Accuracy: ",accuracy_score(target_test,pred,normalize = True ))
```

### K- Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

neigh = KNeighborsClassifier(n_neighbors = 3) #Create object neigh of the type KNeighborsClassifier
neigh.fit(data2_train,target_train)
pred = neigh.predict(data2_test)
print("KNeighbors accuracy score:",accuracy_score(target_test,pred))
```

#### Accuracy: 
Ratio of correctly predicted observation to the total observations 
Accuracy = TP + TN/TP+FP+FN+TN

#### Precision: 
Rati of correctly predicted positive observations to the total predicted positive observation 
Precision = TP/TP+FP

#### Recall:
Also known as sensitivity. Ratio of correctly predicted positive observations to the all observations in actual call. 
Recall = TP/TP+FN


#### F1 Score: 
F1 Score is the weighted average of Precision and Recall.
F1 Score = 2*(Recall* Precision)/ (Recall+Precision)
